```python
import math
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = []
v = [0,0,0,0,0,0]
f = []
p = [0,0,0,0,0,0]
g_bestler = []
```

```python
def v_fonk(x,v,p,g):
    r1 = rnd.random()
    r2 = rnd.random()
    return v + r1*(p-x) + r2*(g-x)
```

```python
def x_fonk(x,v):
    return x + v
```

```python
for i in range(6):
    sayi = rnd.uniform(-10,10)
    x.append(sayi)
```

```python
def fonk(x):
    return math.sin(x)+math.sin(3.33333*x)
```

```python
for i in x:
    f.append(fonk(i))
```

```python
for i in range(6):
        p[i] = x[i]
```

```python
def p_hesapla(x,f,j):
    for i in range(6):
        if(f[i+j]>f[j-6+i]):
            p.append((x[f.index(f[i+j])]))
        else:
            p.append((x[f.index(f[i])]))
```

```python
def g_hesapla(x,f):
    g_bestler.append(x[f.index(max(f))])
```

```python
def f_hesapla(x,j):
    for i in range(j,j+6):
        f.append(fonk(x[i]))
```

```python
g_hesapla(x,f)
j = 0
k = 0
while(j!=6000):
    
    for i in range(6):
        v.append(v_fonk(x[i+j],v[i+j],p[i+j],g_bestler[k]))

    for i in range(6):
        say = x_fonk(x[i+j],v[i+j])
        if((say > 10) | (say < -10)):
            x.append(x[i+j])
        else:
            x.append(say)
        
    p_hesapla(x,f,j)
    f_hesapla(x,j+6)
    g_hesapla(x,f)
    print(g_bestler[k])
    print(fonk(g_bestler[k]))
    print("+++")
    j = j + 6
    k = k + 1

print(x[f.index(max(f))])
print(f.index(max(f)))
print(max(f))
```

```python
plt.figure()
plt.suptitle("Fitness deerlerinin global bestlere gre deiimi")
plt.subplot(121)
plt.plot(g_bestler, 'r')
plt.title("Global Bestler")
plt.subplot(122)
plt.plot(f, 'b')
plt.title("Fitness Deerleri")
plt.show()
```
